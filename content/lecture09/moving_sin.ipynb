```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

%matplotlib ipympl
```

```python
fig, ax = plt.subplots()
ax.set(xlim=(0, 2), ylim=(-2, 2), xlabel='x', ylabel='y', title='Moving sin')
ax.spines['bottom'].set_position('zero')
line, = ax.plot([], [], 'r-', lw=2)
text = ax.text(1.75, 1.5, '0')

x = np.linspace(0, 2, 200)

def repaint(i):
    line.set_data(x, np.sin(2 * np.pi * (x - 0.01 * i)))
    text.set_text(f'{i}')
    return (line,text)

anim = FuncAnimation(fig, repaint, frames=100, interval=20, blit=True)
```
