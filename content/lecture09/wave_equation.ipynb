```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

%matplotlib ipympl
```

```python
def show_wave_solution():
    fig, ax = plt.subplots()
    ax.set(xlim=(-8, 8), ylim=(-1.2, 1.2), xlabel='x', ylabel='y', title='Wave equation solution')
    num_frames = 500
    t_max = 10.0

    line, = ax.plot([], [], 'r-', lw=2)

    x = np.linspace(-8, 8, 500)

    def repaint(i):
        t = t_max * (2*i/num_frames - 1)
        y = np.exp(-(x - t)**2) - 0.5 * np.exp(-(x + t)**2)
        line.set_data(x, y)
        return line,

    anim = FuncAnimation(fig, repaint, frames=num_frames, interval=20, blit=True)
    return anim

wave_anim = show_wave_solution()
```

```python
def show_diffusion_solution():
    fig, ax = plt.subplots()
    ax.set(xlim=(-4, 4), ylim=(0, 4), xlabel='x', ylabel='y', title='Diffusion equation solution')
    num_frames = 200
    t_min = 0.1
    t_max = 10.0

    line, = ax.plot([], [], 'r-', lw=2)

    x = np.linspace(-4, 4, 200)

    def repaint(i):
        t = t_min + (t_max - t_min) * i/num_frames
        y = np.exp(-x ** 2/ (4 * t)) / np.sqrt(t)
        line.set_data(x, y)
        return line,

    anim = FuncAnimation(fig, repaint, frames=num_frames, interval=20, blit=True)
    return anim

diffusion_anim = show_diffusion_solution()
```
