```python
import numpy as np
import matplotlib.pyplot as plt
```

A Lissajous curve is given by parametric equations
$$ x = A \sin(\alpha t+\phi) \hspace{5em} y = B \sin(\beta t) $$
(where $A$, $B$, $\alpha$, $\beta$ and $\phi$ are constants and $t$ is variable).  The most interesting case is when $\alpha$ and $\beta$ are integers.

```python
def plot_lissajous(A = 1, B = 1, alpha = 1, beta = 1, phi = np.pi/2):
    t = np.linspace(0, 4*np.pi, 2000)
    x = A * np.sin(alpha * t + phi)
    y = B * np.sin(beta * t)
    plt.plot(x, y, 'r-')
    plt.axis('equal')
```

```python
plot_lissajous(A = 2,alpha = 3, beta = 3*np.pi)
```
