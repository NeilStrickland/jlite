```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(0, 2*np.pi, 100)  # x coordinates of points on the graph
ys = np.sin(xs)                    # y coordinates of points on the graph
plt.plot(xs, ys)
```
