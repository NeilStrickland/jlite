```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

Here is a nonlinear function, which depends in a nonlinear way on the parameters $a$ to $d$.

```python
def f(x,a,b,c,d):
    return a*np.exp(-b*x)*np.sin(c*x) + d
```

We now make a random dataset in which `y` is approximately (but not exactly) `f(x,a,b,c,d)` for some parameters $a$ to $d$.

```python
N = 100
a = 4 + np.random.random()
b = 1 + np.random.random()
c = 4 + 2 * np.random.random()
d = 6 + 3 * np.random.random()
x = np.linspace(0,5,N)
y0 = f(x,a,b,c,d)
y = y0 + 0.3 * np.random.normal(size=N)
plt.plot(x,y,'.')
```

We now use `scipy.optimize.curvefit()` to try to reconstruct the parameters $a$ to $d$ from the data.

```python
p0 = [4,1,5,8]
p, cov = curve_fit(f, x, y, p0)
print(f'Estimated [a,b,c,d] = {p}')
y1 = f(x,*p)
plt.plot(x,y,'b.')
plt.plot(x,y0,'g-')
plt.plot(x,y1,'r-');
```
