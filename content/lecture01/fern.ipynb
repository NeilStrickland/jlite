```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
F = np.array(
  [[[ 0.00, 0.00, 0.00],[ 0.00, 0.16, 0.00],[ 0.00, 0.00, 1.00]],
   [[ 0.85, 0.04, 0.00],[-0.04, 0.85, 1.60],[ 0.00, 0.00, 1.00]],
   [[ 0.20,-0.26, 0.00],[ 0.23, 0.22, 1.60],[ 0.00, 0.00, 1.00]],
   [[-0.15, 0.28, 0.00],[ 0.26, 0.24, 0.44],[ 0.00, 0.00, 1.00]]]
   )

def next_point(u):
    return F[np.random.choice(range(4),p=[0.02,0.84,0.07,0.07])] @ u

N = 100000
N0 = 1000
x = np.zeros((N,3))
x[0,2] = 1
for i in range(N-1):
    x[i+1] = next_point(x[i])
x = x[N0:]

fig, ax = plt.subplots()
fig.set_size_inches(6,10)
ax.axis('off')
ax.plot(x[:,0], x[:,1], ',', c='g');
```
