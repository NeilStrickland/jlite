```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url = "https://covid19.who.int/WHO-COVID-19-global-data.csv"
df = pd.read_csv(url, parse_dates=['Date_reported'], index_col='Date_reported')
print(df.shape)
df.head()
```

```python
df_UK = df[df.Country_code == 'GB']
print(df_UK.shape)
df_UK.head()
```

```python
plt.figure(figsize=(15, 5))
df_UK['New_cases'].plot(color='red', style='.');
df_UK['New_cases'].rolling(7).mean().plot(color = 'blue');
```
